# Analyzing COVID-19 Data in Python

# Data Source

The data is from the European Centre for Disease Prevention and Control

```python
%matplotlib inline
import pandas
```

```python
df = pandas.read_excel('s3://hamilton-ia241/COVID-19-geographic-disbtribution-worldwide-2020-05-01.xlsx')
df
```

# Average number of cases per countriesAndTerritories

```python
avg_cases_per_countriesAndTerritories = df.groupby('countriesAndTerritories').mean()['cases']

avg_cases_per_countriesAndTerritories
```

# United States of America Cases related to Deaths

```python
United_States_of_America_data = df.loc[df['countriesAndTerritories'] == 'United_States_of_America']
United_States_of_America_data.plot.scatter(x='cases',y='deaths')
```

# The Top 20 Countries with the Highest Deaths

```python
sum_death_by_country = df.groupby('countriesAndTerritories').sum()['deaths']
sum_death_by_country.nlargest(20).plot.bar()
```
